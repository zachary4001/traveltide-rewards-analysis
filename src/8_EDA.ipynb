### User Features table Data Validation

```python
# updated for v2 user_features
# -- BLOCK 1: Row count and column count sanity check
q_block1 = """
SELECT 
    COUNT(*)                                    AS total_users,
    COUNT(DISTINCT user_id)                     AS unique_users,
    SUM(is_never_booked)                        AS never_booked,
    COUNT(CASE WHEN user_booking_status = 'cancelled_all' 
        THEN 1 END)                             AS cancelled_all,
    COUNT(CASE WHEN user_booking_status = 'booked' 
        THEN 1 END)                             AS booked
FROM tt_project.user_features;
"""
# -- BLOCK 2: New distance fields — null counts and distributions
q_block2 = """ 
SELECT
    COUNT(attempted_avg_km)                     AS attempted_avg_km,
    COUNT(attempted_max_km)                     AS attempted_max_km,
    COUNT(attempted_intl_flights)               AS attempted_intl_flights,
    COUNT(attempted_domestic_flights)           AS attempted_domestic_flights,
    COUNT(completed_avg_km)                     AS completed_avg_km,
    COUNT(completed_max_km)                     AS completed_max_km,
    COUNT(completed_intl_flights)               AS completed_intl_flights,
    COUNT(pct_complete_flights_intl)            AS pct_complete_intl_flights,
    COUNT(cxl_avg_km)                           AS cxl_avg_km,
    COUNT(cxl_max_km)                           AS cxl_max_km,
    COUNT(pct_cxl_flights_intl)                 AS pct_cxl_intl_flights,
    -- expected nulls: hotel-only + never-booked (~818)
    5998 - COUNT(completed_avg_km)              AS expected_null_flight_users
FROM tt_project.user_features;
"""
# -- BLOCK 3: New booking type fields — null and sum checks
q_block3 = """
SELECT
    SUM(total_package_bookings)                 AS total_package_bookings,
    SUM(total_flight_only_bookings)             AS total_flight_only_bookings,
    SUM(total_hotel_only_bookings)              AS total_hotel_only_bookings,
    -- package + flight_only + hotel_only should = total_bookings sum
    SUM(total_bookings)                         AS total_bookings,
    SUM(total_package_bookings 
        + total_flight_only_bookings 
        + total_hotel_only_bookings)            AS booking_type_sum,
    -- delta should be 0
    SUM(total_bookings) - SUM(total_package_bookings 
        + total_flight_only_bookings 
        + total_hotel_only_bookings)            AS booking_type_delta
FROM tt_project.user_features;
"""
# -- BLOCK 4: Cancellation booking type math check
q_block4 = """
SELECT
    SUM(total_cancellations)                    AS total_cancellations,
    SUM(package_cancellations 
        + flight_only_cancellations 
        + hotel_only_cancellations)             AS cxl_type_sum,
    -- delta should be 0
    SUM(total_cancellations) - SUM(package_cancellations 
        + flight_only_cancellations 
        + hotel_only_cancellations)             AS cxl_type_delta,
    COUNT(CASE WHEN total_cancellation_rate > 1 
        THEN 1 END)                             AS cxl_rate_over_1,
    MAX(total_cancellation_rate)                AS max_cxl_rate
FROM tt_project.user_features;
"""
# -- BLOCK 5: Logical Consistency - Granular Validation ------------------─
# Updated from original v1 fields + expanded for new fields
# All results should be 0 unless documented exceptions exist
# exception: intl_dom_sum_not_1 - floating point rounding artifacts are acceptable at < 5 users
# exception: attempted_less_than_completed_avg - 184 documented valid cases 
#           (users who cancel shorter trips, complete longer ones)
# exception: booked_has_no_trips - could surface users whose only trips are pre-cohort; 
#           document if non-zero rather than treating as a bug
#
q_block5 = """
SELECT
    -- -- RATE BOUNDS (all should be 0-1) ----------------------------─
    SUM(CASE WHEN booking_conversion_rate > 1 
        THEN 1 ELSE 0 END)                          AS conversion_over_1,
    SUM(CASE WHEN flight_discount_rate > 1 
        THEN 1 ELSE 0 END)                          AS flight_disc_over_1,
    SUM(CASE WHEN hotel_discount_rate > 1 
        THEN 1 ELSE 0 END)                          AS hotel_disc_over_1,
    SUM(CASE WHEN both_discount_rate > 1 
        THEN 1 ELSE 0 END)                          AS both_disc_over_1,
    SUM(CASE WHEN overall_discount_pct > 1 
        THEN 1 ELSE 0 END)                          AS overall_disc_over_1,
    SUM(CASE WHEN avg_flight_discount_pct > 1 
        THEN 1 ELSE 0 END)                          AS flight_disc_pct_over_1,
    SUM(CASE WHEN avg_hotel_discount_pct > 1 
        THEN 1 ELSE 0 END)                          AS hotel_disc_pct_over_1,

    -- -- CANCELLATION RATE BOUNDS --------------------------------------
    SUM(CASE WHEN total_cancellation_rate > 1 
        THEN 1 ELSE 0 END)                          AS total_cxl_rate_over_1,
    SUM(CASE WHEN hotel_only_cxl_rate > 1 
        THEN 1 ELSE 0 END)                          AS hotel_cxl_rate_over_1,
    SUM(CASE WHEN flight_only_cxl_rate > 1 
        THEN 1 ELSE 0 END)                          AS flight_cxl_rate_over_1,
    SUM(CASE WHEN package_cxl_rate > 1 
        THEN 1 ELSE 0 END)                          AS package_cxl_rate_over_1,
    SUM(CASE WHEN cancel_spend_ratio > 1 
        THEN 1 ELSE 0 END)                          AS cancel_ratio_over_1,

    -- -- DISTANCE PCT BOUNDS (all should be 0-1) ----------------------─
    SUM(CASE WHEN pct_complete_flights_intl > 1 
        THEN 1 ELSE 0 END)                          AS pct_comp_intl_over_1,
    SUM(CASE WHEN pct_complete_flights_domestic > 1 
        THEN 1 ELSE 0 END)                          AS pct_comp_dom_over_1,
    SUM(CASE WHEN pct_cxl_flights_intl > 1 
        THEN 1 ELSE 0 END)                          AS pct_cxl_intl_over_1,
    SUM(CASE WHEN pct_cxl_flights_domestic > 1 
        THEN 1 ELSE 0 END)                          AS pct_cxl_dom_over_1,
    -- intl + domestic should sum to 1 for flight bookers
    -- flagging any users where sum deviates by more than 0.001
    SUM(CASE WHEN completed_avg_km IS NOT NULL
        AND ABS((pct_complete_flights_intl 
            + pct_complete_flights_domestic) - 1) > 0.001
        THEN 1 ELSE 0 END)                          AS intl_dom_sum_not_1,

    -- -- SPEND LOGIC --------------------------------------------------─
    SUM(CASE WHEN completed_total_gross > attempted_total_gross 
        THEN 1 ELSE 0 END)                          AS completed_exceeds_attempted,
    SUM(CASE WHEN completed_total_net > completed_total_gross 
        THEN 1 ELSE 0 END)                          AS net_exceeds_gross_completed,
    SUM(CASE WHEN attempted_total_net > attempted_total_gross 
        THEN 1 ELSE 0 END)                          AS net_exceeds_gross_attempted,
    SUM(CASE WHEN cxl_gross_spend > attempted_total_gross 
        THEN 1 ELSE 0 END)                          AS cxl_exceeds_attempted,

    -- -- TRIP COUNT LOGIC ----------------------------------------------
    SUM(CASE WHEN sess_total_trips > life_total_trips 
        THEN 1 ELSE 0 END)                          AS sess_exceeds_life_trips,
    SUM(CASE WHEN total_cancellations > total_booking_sessions 
        THEN 1 ELSE 0 END)                          AS cancels_exceed_bookings,
    -- booking type totals should equal total_bookings
    SUM(CASE WHEN (total_package_bookings 
        + total_flight_only_bookings 
        + total_hotel_only_bookings) != total_bookings
        THEN 1 ELSE 0 END)                          AS booking_type_mismatch,
    -- cancellation type totals should equal total_cancellations
    SUM(CASE WHEN (package_cancellations 
        + flight_only_cancellations 
        + hotel_only_cancellations) != total_cancellations
        THEN 1 ELSE 0 END)                          AS cxl_type_mismatch,

    -- -- DEMOGRAPHIC BOUNDS --------------------------------------------
    SUM(CASE WHEN user_age < 18 OR user_age > 100 
        THEN 1 ELSE 0 END)                          AS impossible_age,
    SUM(CASE WHEN active_tenure_days < 0 
        THEN 1 ELSE 0 END)                          AS negative_tenure,
    SUM(CASE WHEN days_since_signup < 0 
        THEN 1 ELSE 0 END)                          AS negative_days_signup,
    SUM(CASE WHEN signup_to_first_purchase_days < 0 
        THEN 1 ELSE 0 END)                          AS negative_signup_to_purchase,

    -- -- LOYALTY / BEHAVIOR BOUNDS ------------------------------------─
    SUM(CASE WHEN trips_per_year < 0 
        THEN 1 ELSE 0 END)                          AS negative_trips_per_year,
    SUM(CASE WHEN avg_days_between_trips < 0 
        THEN 1 ELSE 0 END)                          AS negative_days_between_trips,
    SUM(CASE WHEN avg_lead_time_days < 0 
        THEN 1 ELSE 0 END)                          AS negative_lead_time,
    SUM(CASE WHEN avg_trip_nights < 0 
        THEN 1 ELSE 0 END)                          AS negative_trip_nights,
    SUM(CASE WHEN avg_checked_bags < 0 
        THEN 1 ELSE 0 END)                          AS negative_checked_bags,

    -- -- DISTANCE LOGIC ------------------------------------------------
    SUM(CASE WHEN completed_avg_km < 0 
        THEN 1 ELSE 0 END)                          AS negative_completed_km,
    SUM(CASE WHEN attempted_avg_km < 0 
        THEN 1 ELSE 0 END)                          AS negative_attempted_km,
    SUM(CASE WHEN cxl_avg_km < 0 
        THEN 1 ELSE 0 END)                          AS negative_cxl_km,
    -- completed_max should always >= completed_avg for flight users
    SUM(CASE WHEN completed_max_km < completed_avg_km 
        THEN 1 ELSE 0 END)                          AS max_less_than_avg_completed,
    SUM(CASE WHEN attempted_max_km < attempted_avg_km 
        THEN 1 ELSE 0 END)                          AS max_less_than_avg_attempted,
    -- never-booked users should have NULL distance, not 0
    SUM(CASE WHEN is_never_booked = 1 
        AND completed_avg_km IS NOT NULL 
        THEN 1 ELSE 0 END)                          AS never_booked_has_distance,

    -- -- BOOKING STATUS LOGIC ------------------------------------------
    -- cancelled_all users should have 0 completed trips
    SUM(CASE WHEN user_booking_status = 'cancelled_all' 
        AND sess_total_trips > 0 
        THEN 1 ELSE 0 END)                          AS cancelled_all_has_trips,
    -- never_booked users should have 0 bookings
    SUM(CASE WHEN user_booking_status = 'never_booked' 
        AND total_bookings > 0 
        THEN 1 ELSE 0 END)                          AS never_booked_has_bookings,
    -- booked users should have at least 1 completed trip
    SUM(CASE WHEN user_booking_status = 'booked' 
        AND sess_total_trips = 0 
        THEN 1 ELSE 0 END)                          AS booked_has_no_trips

FROM tt_project.user_features;
"""
# -- BLOCK 6: Distance logic consistency
# -- attempted >= completed for avg and max
q_block6 = """
SELECT
    COUNT(CASE WHEN attempted_avg_km < completed_avg_km 
        THEN 1 END)                             AS attempted_less_than_completed_avg,
    COUNT(CASE WHEN attempted_max_km < completed_max_km 
        THEN 1 END)                             AS attempted_less_than_completed_max,
    -- pct fields should be between 0 and 1
    COUNT(CASE WHEN pct_complete_flights_intl > 1 
        THEN 1 END)                             AS pct_complete_intl_over_1,
    COUNT(CASE WHEN pct_cxl_flights_intl > 1 
        THEN 1 END)                             AS pct_cxl_intl_over_1,
    -- intl + domestic pct should sum to ~1 for flight bookers
    ROUND(AVG(CASE WHEN completed_avg_km IS NOT NULL 
        THEN pct_complete_flights_intl 
            + pct_complete_flights_domestic END)::numeric, 4) 
                                                AS avg_intl_domestic_sum
FROM tt_project.user_features;
"""
# -- BLOCK 7: High level distance story by booking status
# -- using group averages based on total distance / total flights
q_block7 = """
SELECT
    user_booking_status,
    COUNT(*)                                                    AS user_count,
    ROUND(SUM(attempted_total_km)::numeric
        / NULLIF(SUM(attempted_intl_flights
            + attempted_domestic_flights), 0)
    , 0)                                               AS avg_attempted_km,
    ROUND(SUM(completed_total_km)::numeric
        / NULLIF(SUM(completed_intl_flights
            + completed_domestic_flights), 0)
    , 0)                                               AS avg_completed_km,
    ROUND(SUM(cxl_total_km)::numeric
        / NULLIF(SUM(total_cxl_intl_flights
            + total_cxl_domestic_flights), 0)
    , 0)                                               AS avg_cxl_km,
    -- Pct of attempted intl flights that were completed
    ROUND(SUM(completed_intl_flights)::numeric
        / NULLIF(SUM(attempted_intl_flights), 0)
    , 3)                                               AS pct_intl_completed,
    -- Inverse = pct of attempted intl that were cancelled
    ROUND(SUM(total_cxl_intl_flights)::numeric
        / NULLIF(SUM(attempted_intl_flights), 0)
    , 3)                                               AS pct_intl_cancelled,
    -- Pct of attempted domestic flights that were completed
    ROUND(SUM(completed_domestic_flights)::numeric
        / NULLIF(SUM(attempted_domestic_flights), 0)
    , 3)                                               AS pct_domestic_completed,

    -- Context counts
    COUNT(completed_avg_km)                                     AS users_with_completed_km,
    COUNT(cxl_avg_km)                                           AS users_with_cxl_km

FROM tt_project.user_features
GROUP BY user_booking_status
ORDER BY user_booking_status;
"""
# -- BLOCK 8: High level distance story by booking type mix
q_block8 = """
SELECT
    CASE 
        WHEN total_flight_only_bookings > 0 
            AND total_hotel_only_bookings = 0 
            AND total_package_bookings = 0  THEN 'flight_only'
        WHEN total_hotel_only_bookings > 0 
            AND total_flight_only_bookings = 0 
            AND total_package_bookings = 0  THEN 'hotel_only'
        WHEN total_package_bookings > 0 
            AND total_flight_only_bookings = 0 
            AND total_hotel_only_bookings = 0   THEN 'package_only'
        ELSE 'mixed' 
    END                                         AS booker_type,
    COUNT(*)                                    AS user_count,
    ROUND(AVG(completed_avg_km)::numeric, 0)    AS avg_completed_km,
    ROUND(AVG(pct_complete_flights_intl)
        ::numeric, 3)                           AS avg_pct_intl,
    ROUND(AVG(total_cancellation_rate)
        ::numeric, 3)                           AS avg_cxl_rate
FROM tt_project.user_features
WHERE user_booking_status = 'booked'
GROUP BY booker_type
ORDER BY user_count DESC;
"""
# -- BLOCK 9: Diagnostic eval
q_block9 = """
SELECT
    SUM(ta.count_total_trips)               AS completed_trips,
    SUM(tcx.count_total_trips)              AS cancelled_trips,
    SUM(COALESCE(ta.count_total_trips,0) + 
        COALESCE(tcx.count_total_trips,0))  AS combined_trips,
    SUM(sa.total_bookings)                  AS sa_total_bookings,
    SUM(ta.count_package_trips)             AS completed_package,
    SUM(tcx.count_package_trips)            AS cancelled_package,
    SUM(ta.count_hotel_only_trips)             AS completed_hotel,
    SUM(tcx.count_hotel_only_trips)            AS cancelled_hotel,
    SUM(ta.count_flight_only_trips)             AS completed_flight,
    SUM(tcx.count_flight_only_trips)            AS cancelled_flight,
    SUM(sa.attempted_intl_flights 
        + sa.attempted_domestic_flights)    AS attempted_flights_sum
FROM tt_project.session_agg sa
LEFT JOIN tt_project.trip_agg ta      ON sa.user_id = ta.user_id
LEFT JOIN tt_project.trip_cx_agg tcx  ON sa.user_id = tcx.user_id;
"""
# -- BLOCK 10: Never-booked user profile 
q_block10 = """
SELECT
    COUNT(*)                                            AS user_count,
    ROUND(AVG(completed_total_gross)::numeric, 2)       AS avg_completed_gross,
    ROUND(AVG(attempted_total_gross)::numeric, 2)       AS avg_attempted_gross,
    ROUND(AVG(avg_fare_per_seat)::numeric, 2)           AS avg_fare,
    ROUND(AVG(avg_trip_nights)::numeric, 2)             AS avg_nights,
    ROUND(AVG(total_sessions)::numeric, 1)              AS avg_sessions,
    ROUND(AVG(avg_clicks_per_session)::numeric, 1)      AS avg_clicks,
    SUM(total_cancellations)                            AS total_cancels,
    ROUND(AVG(days_since_signup)::numeric, 0)           AS avg_days_signup
FROM tt_project.user_features
WHERE is_never_booked = 1;
"""
# -- BLOCK 11: Days since signup distribution 
q_block11 = """
SELECT 
    sign_up_date,
    days_since_signup,
    COUNT(*)                                            AS user_count
FROM tt_project.user_features
GROUP BY sign_up_date, days_since_signup
ORDER BY days_since_signup NULLS FIRST
;
"""
# -- BLOCK 12: Cancellation rate outlier check  
q_block12 = """
SELECT
    user_id,
    total_cancellation_rate,
    total_cancellations,
    total_booking_sessions,
    sess_total_trips,
    life_total_trips,
    cxl_gross_spend,
    cancel_spend_ratio
FROM tt_project.user_features
WHERE total_cancellation_rate > 1
ORDER BY total_cancellation_rate DESC
LIMIT 10;
"""
# -- BLOCK 13: Cancel vs completed trip profile comparison 
q_block13 = """
SELECT
    CASE WHEN total_cancellations > 0 
        THEN 'has_cancellations' 
        ELSE 'no_cancellations' END                     AS cancel_profile,
    COUNT(*)                                            AS user_count,
    ROUND(AVG(total_cancellation_rate)::numeric, 3)     AS avg_cancel_rate,
    ROUND(AVG(sess_total_trips)::numeric, 2)            AS avg_completed_trips,
    ROUND(AVG(total_cancellations)::numeric, 2)         AS avg_cancelled_trips,
    ROUND(AVG(completed_total_gross)::numeric, 2)       AS avg_completed_spend,
    ROUND(AVG(cxl_gross_spend)::numeric, 2)             AS avg_cancelled_spend,
    ROUND(AVG(cancel_spend_ratio)::numeric, 3)          AS avg_cancel_spend_ratio,
    ROUND(AVG(avg_cancelled_lead_time)::numeric, 1)     AS avg_cx_lead_time,
    ROUND(AVG(avg_lead_time_days)::numeric, 1)          AS avg_completed_lead_time,
    ROUND(AVG(avg_net_cancelled_flight)::numeric, 2)    AS avg_cx_fare,
    ROUND(AVG(avg_fare_per_seat)::numeric, 2)           AS avg_completed_fare,

    ROUND(SUM(completed_total_km)::numeric
        / NULLIF(SUM(completed_intl_flights
            + completed_domestic_flights), 0)
    , 0)                                               AS avg_completed_km,
    ROUND(SUM(cxl_total_km)::numeric
        / NULLIF(SUM(total_cxl_intl_flights
            + total_cxl_domestic_flights), 0)
    , 0)                                               AS avg_cxl_km,
    -- Pct of attempted intl flights that were cancelled
    ROUND(SUM(total_cxl_intl_flights)::numeric
        / NULLIF(SUM(attempted_intl_flights), 0)
    , 3)                                               AS pct_intl_cancelled,
    -- Pct of attempted domestic flights that were cancelled
    ROUND(SUM(total_cxl_domestic_flights)::numeric
        / NULLIF(SUM(attempted_domestic_flights), 0)
    , 3)                                               AS pct_domestic_cancelled

FROM tt_project.user_features
GROUP BY cancel_profile
ORDER BY cancel_profile;
"""
# -- BLOCK 14: Perk signal preview by booking status  
q_block14 = """
SELECT
    user_booking_status,
    COUNT(*)                                            AS user_count,
    ROUND(AVG(avg_clicks_per_session)::numeric, 1)      AS avg_clicks,
    ROUND(AVG(avg_session_duration_sec)::numeric, 0)    AS avg_duration_sec,
    ROUND(AVG(booking_conversion_rate)::numeric, 3)     AS avg_conversion,
    ROUND(AVG(overall_discount_pct)::numeric, 3)        AS avg_discount_pct,
    ROUND(AVG(flight_discount_rate)::numeric, 3)        AS avg_flight_disc_rate,
    ROUND(AVG(completed_total_gross)::numeric, 2)       AS avg_completed_spend,
    ROUND(AVG(total_cancellation_rate)::numeric, 3)     AS avg_cancel_rate,
    ROUND(AVG(avg_cancelled_lead_time)::numeric, 1)     AS avg_cx_lead_time,
    ROUND(AVG(trips_per_year)::numeric, 2)              AS avg_trips_yr,
    ROUND(AVG(days_since_signup)::numeric, 0)           AS avg_days_signup,
    ROUND(AVG(avg_checked_bags)::numeric, 2)            AS avg_checked_bags,

    -- Weighted avg completed km per group
    ROUND(SUM(completed_total_km)::numeric
        / NULLIF(SUM(completed_intl_flights
            + completed_domestic_flights), 0)
    , 0)                                               AS avg_completed_km,
    -- True pct intl = group intl completed / group total completed flights
    ROUND(SUM(completed_intl_flights)::numeric
        / NULLIF(SUM(completed_intl_flights
            + completed_domestic_flights), 0)
    , 3)                                               AS pct_intl_flights

FROM tt_project.user_features
GROUP BY user_booking_status
ORDER BY user_booking_status;
"""
# -- Validation Block Runner ------------------------------------------─
# Runs all validation queries, displays with headers and comments, exports to CSV

blocks = [
    (
        "q_block1",
        "Block 1:  Basic Structure",
        "Users count validation: with breakout of user booking category"
    ),
    (
        "q_block2",
        "Block 2:  Distance Fields — NULL Counts",
        "Shows count of rows NOT NULL: (hotel-only + never-booked ~818), Users no complete flights ~953"
    ),
    (
        "q_block3",
        "Block 3:  Booking Type Fields — NULL and Sum Check",
        "booking_type_delta should be 0 — package+flight+hotel = total_bookings"
    ),
    (
        "q_block4",
        "Block 4:  Cancellation Booking Types - Math Check",
        "cxl_type_delta=0, cxl_rate_over_1=0, max_cxl_rate<=1"
    ),
    (
        "q_block5",
        "Block 5:  Logical Consistency — Granular Validation",
        "All results should be 0 — except intl_dom_sum_not_1 (rounding <5 ok),\nattempted_less_than_completed_avg (184 documented valid), \nbooked_has_no_trips (pre-cohort trips)"
    ),
    (
        "q_block6",
        "Block 6:  Distance Logic Consistency",
        "attempted_less_than_completed_avg = count of users avg travel attempted \nis lower than avg travel completed (184 is valid), \nall pct fields 0-1, intl+domestic sums to 1"
    ),
    (
        "q_block7",
        "Block 7:  Distance Story by Booking Status",
        "potential signal for - No Change Fees perk"
    ),
    (
        "q_block8",
        "Block 8:  Distance Story by Booking Type Mix",
        "Only users with bookings showing flight distances, intl travel and cancel rate"
    ),
    (
        "q_block9",
        "Block 9:  Diagnostic Eval booking types",
        "Sessions table total_bookings=16099; total booked flights ~13717"
    ),
    (
        "q_block10",
        "Block 10: Never-Booked User Profile",
        "All spend/fare/nights should be NULL — session and click metrics only"
    ),
    (
        "q_block11",
        "Block 11: Days Since Signup Distribution",
        "Earliest signups ~459 days ago; review for outliers or clustering in signup dates"
    ),
    (
        "q_block12",
        "Block 12: Cancellation Rate Outlier Check",
        "Should return 0 rows — any result indicates total_cancellation_rate > 1"
    ),
    (
        "q_block13",
        "Block 13: User Profile by Never Cancelled vs others",
        "No cancellations users will have NaN: Review for differences in profiles"
    ),
    (
        "q_block14",
        "Block 14: Perk Signal Preview by Booking Status",
        "Total users patterns based on: session behavior, discounts, spending, \nbooking & cancellations, and travel patterns."
    ),
]

os.makedirs(r'Q:\scripts\projects\traveltide_analysis\data', exist_ok=True)

results = {}
for query_name, label, guidance in blocks:
    print(f"\n ----- {label} ----- ")
    print(f" Guidance--> {guidance}")
    df_result = pd.read_sql(sa.text(eval(query_name)), connection)
    display(df_result)
    results[query_name] = df_result
    safe_label = label.replace(":", "_").replace(" ", "_").replace("—", "_").replace("__", "_")
    df_result.to_csv(
        rf'Q:\scripts\projects\traveltide_analysis\data\{safe_label}.csv',
        index=True
    )
    print(f" ----- {query_name} Exported -----")

print("\n✓ All blocks complete and exported.")
```

```python
# ============================================================
# Step 1c: Log Transforms 
# Applied after capping, before binning
# log1p used throughout — handles zeros safely: log1p(0) = 0
# New log_ columns created alongside originals
# Original values preserved for reference and perk scoring
# ============================================================

log_cols = [
    # Fare averages
    'avg_fare_per_seat',
    'net_avg_fare_per_seat',
    # Hotel per room
    'avg_hotel_per_room',
    'net_avg_hotel_per_room',
    # Hotel total per stay
    'avg_hotel_total_per_stay',
    'net_avg_hotel_total_per_stay',
    # Completed spend totals
    'completed_total_gross',
    'completed_total_net',
    'completed_flight_gross',
    'completed_hotel_gross',
    # Attempted spend totals
    'attempted_total_gross',
    'attempted_total_net',
    'attempted_flight_gross',
    'attempted_hotel_gross',
    # Cancellation spend
    'cxl_gross_spend',
    'avg_net_cancelled_flight',
    'avg_net_cancelled_hotel',
    # Time fields
    'avg_lead_time_days',
    'avg_cancelled_lead_time',
    'active_tenure_days',
    'signup_to_first_purchase_days',
    'avg_days_between_trips',
    # Distance
    'completed_avg_km',
    'attempted_avg_km',
]

print("=== Applying Log1p Transforms ===\n")
for col in log_cols:
    if col in df_users.columns:
        new_col = f'log_{col}'
        df_users[new_col] = np.log1p(df_users[col].fillna(0))
        print(f"{col:<45} -> {new_col}")
    else:
        print(f"WARNING: {col} not found in df_users")

print(f"\ndf_users shape after transforms: {df_users.shape}")
print("\n✓ Step 1c complete — log transforms applied")
```

### Bin creation

```python
# ============================================================
# Step 1d: Behavioral Bins 
# Bins are the PRIMARY feature for session behavior and time fields
# Encoded as integers for clustering compatibility
# Original and bin label columns preserved alongside encoded versions
# ============================================================

# -- Define bins 
bin_definitions = {
    'avg_lead_time_days': {
        'col':    'avg_lead_time_days',
        'bins':   [0, 7, 14, 60, float('inf')],
        'labels': ['last_minute', 'short_notice',
                   'standard', 'advanced'],
        'new':    'lead_time_bin',
        'enc':    {'last_minute': 0, 'short_notice': 1,
                   'standard': 2, 'advanced': 3},
    },
    'avg_cancelled_lead_time': {
        'col':    'avg_cancelled_lead_time',
        'bins':   [-0.001, 0, 7, 14, 60, float('inf')],
        'labels': ['none', 'last_minute', 'short_notice',
                   'standard', 'advanced'],
        'new':    'cancelled_lead_time_bin',
        'enc':    {'none': -1, 'last_minute': 0, 'short_notice': 1,
                   'standard': 2, 'advanced': 3},
    },
    'booking_conversion_rate': {
        'col':    'booking_conversion_rate',
        'bins':   [-0.001, 0, 0.25, 0.50, float('inf')],
        'labels': ['never', 'low', 'moderate', 'high'],
        'new':    'conversion_bin',
        'enc':    {'never': 0, 'low': 1, 'moderate': 2, 'high': 3},
    },
}

# -- Apply bins and encode --------------------------------------------─
print("=== Applying Bins ===\n")
for name, defn in bin_definitions.items():
    df_users[defn['new']] = pd.cut(
        df_users[defn['col']],
        bins=defn['bins'],
        labels=defn['labels'],
        include_lowest=True
    )
    enc_col = f"{defn['new']}_enc"
    df_users[enc_col] = df_users[defn['new']].map(defn['enc'])
    df_users[enc_col] = pd.to_numeric(
        df_users[enc_col], errors='coerce'
    )
    print(f"=== {defn['new']} ===")
    print(df_users[defn['new']].value_counts().sort_index())
    print(f"Nulls in enc: {df_users[enc_col].isna().sum()}")
    print()

# -- Override: pre-cohort bookers with zero cohort conversion ----------
n_override = (
    (df_users['conversion_bin'] == 'never') &
    (df_users['user_booking_status'] == 'booked')
).sum()

df_users.loc[
    (df_users['conversion_bin'] == 'never') &
    (df_users['user_booking_status'] == 'booked'),
    'conversion_bin'
] = 'low'

df_users['conversion_bin_enc'] = df_users['conversion_bin'].map(
    {'never': 0, 'low': 1, 'moderate': 2, 'high': 3}
)
df_users['conversion_bin_enc'] = pd.to_numeric(
    df_users['conversion_bin_enc'], errors='coerce'
)

print(f"=== conversion_bin after override "
      f"({n_override} users corrected) ===")
print(df_users['conversion_bin'].value_counts().sort_index())
print(f"Nulls in enc: {df_users['conversion_bin_enc'].isna().sum()}")
print(f"\n✓ never should = 434 (true never-booked only)")

print(f"\ndf_users shape after bins: {df_users.shape}")
print("\n✓ Step 1d complete — bins applied and validated")
```

## Step 2: PCA + Scree Plot & Transformed Table creation

#### 1st pass with currated list

```python
# -- Step 2: PCA + Scree Plot ------------------------------------------
# Validates dimensionality of the data before clustering
# Confirms whether 3-dimension approach is appropriate
# Identifies which features drive each component

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -- 1. Define candidate features for PCA ----------------------------─
# Using log-transformed versions where available
# Excluding: identity, dates, strings, booleans, bin labels
# Excluding: rate/pct fields already bounded 0-1 (no transform needed
#            but including discount rates as they carry signal)
# Note: NaN-containing fields will be filled with 0 for never-bookers
#       or median for structural nulls (hotel-only users)

pca_features = [
    # -- WEBSITE USAGE --------------------------------------------─
    'total_sessions',
    'total_bookings',
    'booking_conversion_rate',
    'log_avg_fare_per_seat',         # using log version
    'avg_clicks_per_session',        # continuous, no log needed
    'avg_session_duration_sec',      # continuous, no log needed

    # -- SPENDING --------------------------------------------------
    'log_avg_fare_per_seat',
    'log_avg_hotel_per_room',
    'log_avg_hotel_total_per_stay',
    'log_completed_total_gross',
    'log_attempted_total_gross',
    'flight_discount_rate',
    'hotel_discount_rate',
    'overall_discount_pct',
    'discount_dependency',

    # -- TRAVEL PATTERNS ------------------------------------------─
    'avg_trip_nights',
    'avg_checked_bags',
    'avg_seats',
    'return_flight_rate',
    'destination_variety',
    'airline_variety',
    'lead_time_bin_enc',
    'log_avg_lead_time_days',

    # -- CANCELLATION ----------------------------------------------
    'total_cancellation_rate',
    'cancel_spend_ratio',
    'log_avg_cancelled_lead_time',
    'cancelled_lead_time_bin_enc',

    # -- DISTANCE / INTERNATIONAL ----------------------------------
    'log_completed_avg_km',
    'pct_complete_flights_intl',
    'pct_complete_flights_domestic',

    # -- LOYALTY --------------------------------------------------─
    'life_total_trips',
    'trips_per_year',
    'log_avg_days_between_trips',
    'log_active_tenure_days',
    'log_signup_to_first_purchase_days',
    'family_seats_ratio',
]

# Deduplicate in case any field listed twice
pca_features = list(dict.fromkeys(pca_features))

# -- 2. Build feature matrix ------------------------------------------─
X_pca = df_users[pca_features].copy()

# Fill NaNs:
# Structural nulls (hotel-only, never-booked) → 0
# Remaining nulls → median (avoids distorting scale)
X_pca = X_pca.fillna(0)

print(f"PCA feature matrix: {X_pca.shape}")
print(f"Any remaining nulls: {X_pca.isnull().sum().sum()}")

# -- 3. Scale ----------------------------------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_pca)

# -- 4. Fit PCA — full components ------------------------------------─
pca_full = PCA(random_state=53)
pca_full.fit(X_scaled)

explained = pca_full.explained_variance_ratio_
cumulative = np.cumsum(explained)
n_components = len(explained)

# -- 5. Scree plot ----------------------------------------------------─
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('PCA Scree Plot — Full Feature Set', fontsize=14)

# Individual variance explained
axes[0].bar(range(1, n_components + 1), explained * 100,
            color='steelblue', alpha=0.7, edgecolor='none')
axes[0].plot(range(1, n_components + 1), explained * 100,
             'o-', color='darkblue', markersize=4, linewidth=1.5)
axes[0].axhline(y=1, color='red', linestyle='--', linewidth=1,
                label='1% variance threshold')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Variance Explained (%)')
axes[0].set_title('Individual Variance Explained')
axes[0].set_xlim(0.5, min(25, n_components) + 0.5)
axes[0].legend()

# Cumulative variance explained
axes[1].plot(range(1, n_components + 1), cumulative * 100,
             'o-', color='darkorange', markersize=4, linewidth=1.5)
axes[1].axhline(y=70, color='green', linestyle='--',
                linewidth=1, label='70% threshold')
axes[1].axhline(y=80, color='blue', linestyle='--',
                linewidth=1, label='80% threshold')
axes[1].axhline(y=90, color='red', linestyle='--',
                linewidth=1, label='90% threshold')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Variance Explained (%)')
axes[1].set_title('Cumulative Variance Explained')
axes[1].set_xlim(0.5, min(25, n_components) + 0.5)
axes[1].legend()

plt.tight_layout()
plt.show()

# -- 6. Summary table --------------------------------------------------
print("\n=== Variance Explained by Component ===\n")
print(f"{'Component':<12} {'Individual %':<16} {'Cumulative %':<16}")
print("-" * 44)
for i, (ind, cum) in enumerate(zip(explained[:20], cumulative[:20]), 1):
    marker = " <-- elbow?" if i > 1 and explained[i-2] - ind > 0.02 else ""
    print(f"{i:<12} {ind*100:<16.2f} {cum*100:<16.2f}{marker}")

print(f"\nComponents needed for 70% variance: "
      f"{np.argmax(cumulative >= 0.70) + 1}")
print(f"Components needed for 80% variance: "
      f"{np.argmax(cumulative >= 0.80) + 1}")
print(f"Components needed for 90% variance: "
      f"{np.argmax(cumulative >= 0.90) + 1}")

# -- 7. Top feature loadings per component ----------------------------
print("\n=== Top 5 Features per Component (absolute loading) ===\n")
loadings = pd.DataFrame(
    pca_full.components_[:5].T,
    index=pca_features,
    columns=[f'PC{i+1}' for i in range(5)]
)
for pc in loadings.columns:
    top5 = loadings[pc].abs().sort_values(ascending=False).head(5)
    print(f"{pc}:")
    for feat, val in top5.items():
        direction = "+" if loadings.loc[feat, pc] > 0 else "-"
        print(f"  {direction}{val:.3f}  {feat}")
    print()
```

### Find full list PCA eligible fields

```python
# -- Full eligible feature set for PCA --------------------------------
# Rules for inclusion:
# INCLUDE: all numeric fields that represent user behavior
# EXCLUDE: identity keys, dates, strings, boolean demographics,
#          bin label columns (use encoded versions instead),
#          fields that are mathematical derivatives of others
#          already included (avoid double-counting)

# Fields to explicitly exclude
exclude = set([
    # Identity
    'user_id', 'is_never_booked',
    # Dates (non-numeric after dtype fix)
    'sign_up_date', 'first_session_date',
    'last_session_date', 'first_purchase_date',
    # Demographics — passive, not behavioral
    'user_age', 'married', 'has_children',
    # Bin label columns — use _enc versions instead
    'lead_time_bin', 'cancelled_lead_time_bin', 'conversion_bin',
    # String columns
    'user_booking_status', 'gender',
    'home_country', 'home_city', 'home_airport',
    # Gross/net pairs — use one to avoid double-counting
    # Keep log_ versions where they exist, drop raw where log exists
    # (handled below programmatically)
    # Max distance fields — avg is more stable for clustering
    'completed_max_km', 'attempted_max_km',
    'cxl_max_km',
    # Total distance km — avg already captures this per trip
    'completed_total_km', 'attempted_total_km', 'cxl_total_km',
    # Raw versions where log_ exists — prefer log
    'avg_fare_per_seat', 'net_avg_fare_per_seat',
    'avg_hotel_per_room', 'net_avg_hotel_per_room',
    'avg_hotel_total_per_stay', 'net_avg_hotel_total_per_stay',
    'completed_total_gross', 'completed_total_net',
    'attempted_total_gross', 'attempted_total_net',
    'attempted_flight_gross', 'attempted_hotel_gross',
    'completed_flight_gross', 'completed_hotel_gross',
    'cxl_gross_spend',
    'avg_net_cancelled_flight', 'avg_net_cancelled_hotel',
    'avg_lead_time_days', 'avg_cancelled_lead_time',
    'active_tenure_days', 'signup_to_first_purchase_days',
    'avg_days_between_trips',
    'completed_avg_km', 'attempted_avg_km',
    # Redundant loyalty fields
    'sess_total_trips',     # same as life_total_trips direction
    'avg_cancelled_nights', # sparse, low signal
])

# Build full eligible list programmatically
numeric_cols = df_users.select_dtypes(
    include=['float64', 'int64']
).columns.tolist()

# Add log transforms for discount dollar fields if created
# Remove raw where log version exists
log_cols_present = [c for c in df_users.columns
                    if c.startswith('log_')]
raw_with_log = [c.replace('log_', '') for c in log_cols_present]

# Final eligible set
full_pca_features = [
    c for c in numeric_cols
    if c not in exclude
    and c not in raw_with_log  # prefer log version
] + log_cols_present

# Remove any duplicates
full_pca_features = list(dict.fromkeys(full_pca_features))

print(f"Full eligible feature count: {len(full_pca_features)}")
print(f"\nFeatures included:")
for i, f in enumerate(sorted(full_pca_features), 1):
    print(f"  {i:>3}. {f}")
```

```python
# -- Step 2b: PCA + Scree — Full Clean Feature Set --------------------
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -- 1. Build feature matrix ------------------------------------------─
X_pca = df_users[full_pca_features_clean].copy()

# Fill NaNs:
# Structural nulls (hotel-only, never-booked) → 0
# Remaining → 0 (consistent with zero-fill philosophy)
X_pca = X_pca.fillna(0)

print(f"PCA feature matrix: {X_pca.shape}")
print(f"Any remaining nulls: {X_pca.isnull().sum().sum()}")
print(f"Features: {full_pca_features_clean}")

# -- 2. Scale ----------------------------------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_pca)

# -- 3. Fit PCA — full components --------------------------------------
pca_full = PCA(random_state=53)
pca_full.fit(X_scaled)

explained = pca_full.explained_variance_ratio_
cumulative = np.cumsum(explained)
n_components = len(explained)

# -- 4. Scree plot ----------------------------------------------------─
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle(f'PCA Scree Plot — Full Clean Feature Set '
             f'({len(full_pca_features_clean)} features)',
             fontsize=14)

# Individual variance explained
axes[0].bar(range(1, n_components + 1), explained * 100,
            color='steelblue', alpha=0.7, edgecolor='none')
axes[0].plot(range(1, n_components + 1), explained * 100,
             'o-', color='darkblue', markersize=4, linewidth=1.5)
axes[0].axhline(y=1, color='red', linestyle='--',
                linewidth=1, label='1% variance threshold')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Variance Explained (%)')
axes[0].set_title('Individual Variance Explained')
axes[0].set_xlim(0.5, min(30, n_components) + 0.5)
axes[0].legend()

# Cumulative variance explained
axes[1].plot(range(1, n_components + 1), cumulative * 100,
             'o-', color='darkorange', markersize=4, linewidth=1.5)
for thresh, color, label in [
    (0.70, 'green', '70%'),
    (0.80, 'blue',  '80%'),
    (0.90, 'red',   '90%')
]:
    axes[1].axhline(y=thresh * 100, color=color,
                    linestyle='--', linewidth=1,
                    label=f'{label} threshold')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Variance Explained (%)')
axes[1].set_title('Cumulative Variance Explained')
axes[1].set_xlim(0.5, min(30, n_components) + 0.5)
axes[1].legend()

plt.tight_layout()
plt.show()

# -- 5. Variance table ------------------------------------------------─
print("\n=== Variance Explained by Component ===\n")
print(f"{'Component':<12} {'Individual %':<16} {'Cumulative %':<16} {'Flag'}")
print("-" * 56)
for i, (ind, cum) in enumerate(
        zip(explained[:20], cumulative[:20]), 1):
    drop = explained[i-2] - ind if i > 1 else 0
    flag = " <-- elbow" if i > 1 and drop > 0.03 else ""
    print(f"{i:<12} {ind*100:<16.2f} {cum*100:<16.2f}{flag}")

for thresh, pct in [(0.70, '70%'), (0.80, '80%'), (0.90, '90%')]:
    n = np.argmax(cumulative >= thresh) + 1
    print(f"\nComponents needed for {pct} variance: {n}")

# -- 6. Component loadings heatmap ------------------------------------─
n_show = min(10, n_components)
loadings = pd.DataFrame(
    pca_full.components_[:n_show].T,
    index=full_pca_features_clean,
    columns=[f'PC{i+1}' for i in range(n_show)]
)

fig, ax = plt.subplots(figsize=(16, len(full_pca_features_clean) * 0.35 + 2))
im = ax.imshow(loadings.values, cmap='RdBu_r',
               vmin=-0.5, vmax=0.5, aspect='auto')
ax.set_xticks(range(n_show))
ax.set_xticklabels([f'PC{i+1}' for i in range(n_show)], fontsize=10)
ax.set_yticks(range(len(full_pca_features_clean)))
ax.set_yticklabels(full_pca_features_clean, fontsize=8)
ax.set_title('PCA Component Loadings — Top 10 Components\n'
             '(Red = positive, Blue = negative)', fontsize=12)
plt.colorbar(im, ax=ax, shrink=0.5)
plt.tight_layout()
plt.show()

# -- 7. Top feature loadings per component ----------------------------─
print("\n=== Top 6 Features per Component (absolute loading) ===\n")
for pc_idx in range(min(8, n_components)):
    pc_label = f'PC{pc_idx + 1}'
    top = loadings[pc_label].abs().sort_values(
        ascending=False).head(6)
    pct = explained[pc_idx] * 100
    print(f"{pc_label} ({pct:.1f}% variance):")
    for feat, val in top.items():
        direction = "+" if loadings.loc[feat, pc_label] > 0 else "-"
        print(f"  {direction}{val:.3f}  {feat}")
    print()
```

## Step 4: Elbow & Silhouette - Both Paths

```python
# ── Step 4: Elbow + Silhouette — All Dimensions Both Paths ───────────
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ── Configuration ─────────────────────────────────────────────────────
K_RANGE = range(2, 11)       # test k=2 through k=10
RANDOM_STATE = 53
N_INIT = 10
SAMPLE_SIZE = 2000           # silhouette sample for speed

# ── Combine both paths for single loop ────────────────────────────────
all_dimensions = {
    # Path A
    'A_website':      path_a['website'],
    'A_spending':     path_a['spending'],
    'A_travel':       path_a['travel'],
    'A_cancellation': path_a['cancellation'],
    # Path B
    'B_spending_distance': path_b['spending_distance'],
    'B_cancellation':      path_b['cancellation'],
    'B_hotel_behavior':    path_b['hotel_behavior'],
    'B_flight_discounts':  path_b['flight_discounts'],
    'B_intl_travel':       path_b['intl_travel'],
}

# ── Storage ───────────────────────────────────────────────────────────
results = {}

# ── Run elbow + silhouette per dimension ─────────────────────────────
print("=== Running Elbow + Silhouette Analysis ===\n")
print(f"k range: {list(K_RANGE)}")
print(f"random_state: {RANDOM_STATE}")
print(f"silhouette sample_size: {SAMPLE_SIZE}\n")

for dim_name, features in all_dimensions.items():
    print(f"Processing {dim_name} ({len(features)} features)...",
          end=' ', flush=True)

    # Build and scale feature matrix
    X = df_users[features].fillna(0)
    X_scaled = StandardScaler().fit_transform(X)

    inertias = []
    silhouettes = []

    for k in K_RANGE:
        km = KMeans(n_clusters=k, random_state=RANDOM_STATE,
                    n_init=N_INIT)
        labels = km.fit_predict(X_scaled)
        inertias.append(km.inertia_)
        sil = silhouette_score(X_scaled, labels,
                               sample_size=SAMPLE_SIZE,
                               random_state=RANDOM_STATE)
        silhouettes.append(sil)

    results[dim_name] = {
        'features':    features,
        'inertias':    inertias,
        'silhouettes': silhouettes,
        'best_k_sil':  K_RANGE[np.argmax(silhouettes)],
        'best_sil':    max(silhouettes),
    }
    print(f"best k={results[dim_name]['best_k_sil']} "
          f"(silhouette={results[dim_name]['best_sil']:.4f})")

# ── Plot — Path A ─────────────────────────────────────────────────────
path_a_dims = [d for d in all_dimensions if d.startswith('A_')]
fig, axes = plt.subplots(len(path_a_dims), 2,
                         figsize=(14, len(path_a_dims) * 4))
fig.suptitle('Path A — Elbow + Silhouette per Dimension',
             fontsize=14, y=1.01)

for i, dim_name in enumerate(path_a_dims):
    r = results[dim_name]
    k_list = list(K_RANGE)
    label = dim_name.replace('A_', '')

    # Elbow
    axes[i, 0].plot(k_list, r['inertias'], 'o-',
                    color='steelblue', linewidth=2)
    axes[i, 0].set_title(f'{label} — Inertia (Elbow)')
    axes[i, 0].set_xlabel('k')
    axes[i, 0].set_ylabel('Inertia')
    axes[i, 0].set_xticks(k_list)

    # Silhouette
    axes[i, 1].plot(k_list, r['silhouettes'], 'o-',
                    color='darkorange', linewidth=2)
    axes[i, 1].axvline(r['best_k_sil'], color='red',
                       linestyle='--', linewidth=1.5,
                       label=f"best k={r['best_k_sil']}")
    axes[i, 1].set_title(f'{label} — Silhouette Score')
    axes[i, 1].set_xlabel('k')
    axes[i, 1].set_ylabel('Silhouette')
    axes[i, 1].set_xticks(k_list)
    axes[i, 1].legend()

plt.tight_layout()
plt.show()

# ── Plot — Path B ─────────────────────────────────────────────────────
path_b_dims = [d for d in all_dimensions if d.startswith('B_')]
fig, axes = plt.subplots(len(path_b_dims), 2,
                         figsize=(14, len(path_b_dims) * 4))
fig.suptitle('Path B — Elbow + Silhouette per Dimension',
             fontsize=14, y=1.01)

for i, dim_name in enumerate(path_b_dims):
    r = results[dim_name]
    k_list = list(K_RANGE)
    label = dim_name.replace('B_', '')

    axes[i, 0].plot(k_list, r['inertias'], 'o-',
                    color='steelblue', linewidth=2)
    axes[i, 0].set_title(f'{label} — Inertia (Elbow)')
    axes[i, 0].set_xlabel('k')
    axes[i, 0].set_ylabel('Inertia')
    axes[i, 0].set_xticks(k_list)

    axes[i, 1].plot(k_list, r['silhouettes'], 'o-',
                    color='darkorange', linewidth=2)
    axes[i, 1].axvline(r['best_k_sil'], color='red',
                       linestyle='--', linewidth=1.5,
                       label=f"best k={r['best_k_sil']}")
    axes[i, 1].set_title(f'{label} — Silhouette Score')
    axes[i, 1].set_xlabel('k')
    axes[i, 1].set_ylabel('Silhouette')
    axes[i, 1].set_xticks(k_list)
    axes[i, 1].legend()

plt.tight_layout()
plt.show()

# ── Summary table ─────────────────────────────────────────────────────
print("\n=== Summary — Best k by Silhouette Score ===\n")
print(f"{'Dimension':<25} {'Path':<8} {'Features':<10} "
      f"{'Best k':<8} {'Silhouette':<12} {'Rating'}")
print("-" * 75)

for dim_name, r in results.items():
    path = dim_name[0]
    label = dim_name[2:]
    sil = r['best_sil']
    rating = ('Excellent' if sil > 0.50 else
              'Good'      if sil > 0.35 else
              'Fair'      if sil > 0.25 else
              'Weak')
    print(f"{label:<25} {path:<8} {len(r['features']):<10} "
          f"{r['best_k_sil']:<8} {sil:<12.4f} {rating}")
```

#### Deep dive on Both Paths

```python
# ── Targeted Cluster Analysis — All Dimensions ───────────────────────
# Runs canceller-style targeted analysis per dimension
# Separates structural zero populations before evaluating k
# Returns best k per dimension with full diagnostic output

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def analyze_dimension(name, features, df, k_range=range(2, 9),
                      zero_field=None, zero_label='zero_population',
                      random_state=53, min_cluster_size=30):
    """
    Runs targeted elbow + silhouette on a dimension.
    If zero_field provided, separates zero population first
    and runs clustering only on meaningful population.
    Returns best k based on silhouette, respecting min_cluster_size.
    """
    print(f"\n{'='*60}")
    print(f"  {name} ({len(features)} features)")
    print(f"{'='*60}")

    # Identify analysis population
    if zero_field:
        mask = df[zero_field] > 0
        df_analysis = df[mask].copy()
        n_zero = (~mask).sum()
        print(f"  Zero population ({zero_label}): {n_zero:,} users")
        print(f"  Analysis population: {len(df_analysis):,} users")
    else:
        df_analysis = df.copy()
        n_zero = 0
        print(f"  Full population: {len(df_analysis):,} users")

    X = df_analysis[features].fillna(0)
    X_scaled = StandardScaler().fit_transform(X)

    inertias, silhouettes, size_warnings = [], [], []

    for k in k_range:
        km = KMeans(n_clusters=k, random_state=random_state,
                    n_init=10)
        labels = km.fit_predict(X_scaled)
        inertias.append(km.inertia_)

        sizes = pd.Series(labels).value_counts().sort_values()
        min_size = sizes.min()
        warn = min_size < min_cluster_size

        if len(df_analysis) >= 100:
            sil = silhouette_score(X_scaled, labels)
        else:
            sil = silhouette_score(X_scaled, labels)

        silhouettes.append(sil)
        size_warnings.append(warn)

        warn_str = f" ⚠ min cluster={min_size}" if warn else ""
        print(f"  k={k}: sil={sil:.4f}  "
              f"sizes={sorted(sizes.values.tolist(), reverse=True)}"
              f"{warn_str}")

    # Best k — highest silhouette without small cluster warning
    valid_sils = [s if not w else -1
                  for s, w in zip(silhouettes, size_warnings)]
    best_idx = np.argmax(valid_sils)
    # Fallback to unconstrained best if all have warnings
    if max(valid_sils) == -1:
        best_idx = np.argmax(silhouettes)
        print(f"  ⚠ All k values have small clusters — "
              f"using unconstrained best")

    best_k = list(k_range)[best_idx]
    best_sil = silhouettes[best_idx]
    print(f"\n  → Recommended k={best_k} "
          f"(silhouette={best_sil:.4f})")
    if n_zero > 0:
        print(f"  → Final labels: 0='{zero_label}' + "
              f"{best_k} sub-groups")

    # Plot
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    fig.suptitle(f'{name} — Targeted Analysis'
                 f'{f" (excl. {zero_label}, n={len(df_analysis)})" if zero_field else f" (n={len(df_analysis)})"}',
                 fontsize=11)

    axes[0].plot(list(k_range), inertias, 'o-',
                 color='steelblue', linewidth=2)
    axes[0].set_title('Inertia (Elbow)')
    axes[0].set_xlabel('k')
    axes[0].set_ylabel('Inertia')
    axes[0].set_xticks(list(k_range))

    axes[1].plot(list(k_range), silhouettes, 'o-',
                 color='darkorange', linewidth=2)
    for i, (k, warn) in enumerate(
            zip(k_range, size_warnings)):
        if warn:
            axes[1].plot(k, silhouettes[i], 'rx',
                         markersize=10, linewidth=2)
    axes[1].axvline(best_k, color='red', linestyle='--',
                    linewidth=1.5,
                    label=f'recommended k={best_k}')
    axes[1].set_title('Silhouette (✕ = small cluster warning)')
    axes[1].set_xlabel('k')
    axes[1].set_ylabel('Silhouette')
    axes[1].set_xticks(list(k_range))
    axes[1].legend()

    plt.tight_layout()
    plt.show()

    return {
        'name': name,
        'best_k': best_k,
        'best_sil': best_sil,
        'n_analysis': len(df_analysis),
        'n_zero': n_zero,
        'zero_label': zero_label,
        'inertias': inertias,
        'silhouettes': silhouettes,
    }


# ── Define zero_field per dimension ──────────────────────────────────
# Fields that indicate structural zero populations
# None = run on full cohort

dimension_configs = {
    # PATH A
    'A_website': {
        'features':    path_a['website'],
        'zero_field':  None,
    },
    'A_spending': {
        'features':    path_a['spending'],
        'zero_field':  'completed_total_gross',
        'zero_label':  'non_spender',
    },
    'A_travel': {
        'features':    path_a['travel'],
        'zero_field':  'completed_total_gross',
        'zero_label':  'non_traveler',
    },
    'A_cancellation': {
        'features':    path_a['cancellation'],
        'zero_field':  'total_cancellation_rate',
        'zero_label':  'non_canceller',
    },
    # PATH B
    'B_spending_distance': {
        'features':    path_b['spending_distance'],
        'zero_field':  'completed_total_gross',
        'zero_label':  'non_spender',
    },
    'B_cancellation': {
        'features':    path_b['cancellation'],
        'zero_field':  'total_cancellation_rate',
        'zero_label':  'non_canceller',
    },
    'B_hotel_behavior': {
        'features':    path_b['hotel_behavior'],
        'zero_field':  'total_hotel_only_bookings',
        'zero_label':  'no_hotel_only',
    },
    'B_flight_discounts': {
        'features':    path_b['flight_discounts'],
        'zero_field':  'flight_discount_rate',
        'zero_label':  'no_flight_discount',
    },
    'B_intl_travel': {
        'features':    path_b['intl_travel'],
        'zero_field':  'pct_complete_flights_intl',
        'zero_label':  'domestic_only',
    },
}

# ── Run all dimensions ────────────────────────────────────────────────
dimension_results = {}

for dim_name, config in dimension_configs.items():
    dimension_results[dim_name] = analyze_dimension(
        name=dim_name,
        features=config['features'],
        df=df_users,
        zero_field=config.get('zero_field'),
        zero_label=config.get('zero_label', 'zero_population'),
        min_cluster_size=30,
    )

# ── Final summary ─────────────────────────────────────────────────────
print("\n" + "="*65)
print("  FINAL K RECOMMENDATIONS — ALL DIMENSIONS")
print("="*65)
print(f"\n{'Dimension':<25} {'Path':<6} {'k':<5} "
      f"{'Silhouette':<12} {'N analyzed':<12} {'N zero'}")
print("-" * 65)

for dim_name, r in dimension_results.items():
    path = dim_name[0]
    label = dim_name[2:]
    rating = ('Excellent' if r['best_sil'] > 0.50 else
              'Good'      if r['best_sil'] > 0.35 else
              'Fair'      if r['best_sil'] > 0.25 else 'Weak')
    print(f"{label:<25} {path:<6} {r['best_k']:<5} "
          f"{r['best_sil']:<12.4f} {r['n_analysis']:<12,} "
          f"{r['n_zero']:,}  {rating}")
```

## Step 5: Fit KMeans

```python
# ── Step 5: Fit Final KMeans Models — Both Paths ─────────────────────
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

RANDOM_STATE = 53
N_INIT = 10

# ── Final k values ────────────────────────────────────────────────────
k_final = {
    'A_website':           2,
    'A_spending':          3,
    'A_travel':            2,
    'A_cancellation':      5,
    'B_spending_distance': 2,
    'B_cancellation':      5,
    'B_hotel_behavior':    2,
    'B_flight_discounts':  2,
    'B_intl_travel':       3,
}

# ── Feature maps ──────────────────────────────────────────────────────
feature_map = {
    'A_website':           path_a['website'],
    'A_spending':          path_a['spending'],
    'A_travel':            path_a['travel'],
    'A_cancellation':      path_a['cancellation'],
    'B_spending_distance': path_b['spending_distance'],
    'B_cancellation':      path_b['cancellation'],
    'B_hotel_behavior':    path_b['hotel_behavior'],
    'B_flight_discounts':  path_b['flight_discounts'],
    'B_intl_travel':       path_b['intl_travel'],
}

# ── Zero population handling ──────────────────────────────────────────
# Dimensions with structural zero populations are fit on
# active population only; zero users get label 0
zero_config = {
    'A_spending':          ('completed_total_gross',  'non_spender'),
    'A_travel':            ('completed_total_gross',  'non_traveler'),
    'A_cancellation':      ('total_cancellation_rate','non_canceller'),
    'B_spending_distance': ('completed_total_gross',  'non_spender'),
    'B_cancellation':      ('total_cancellation_rate','non_canceller'),
    'B_hotel_behavior':    ('total_hotel_only_bookings','no_hotel_only'),
    'B_flight_discounts':  ('flight_discount_rate',   'no_flight_discount'),
    'B_intl_travel':       ('pct_complete_flights_intl','domestic_only'),
}

# ── Storage ───────────────────────────────────────────────────────────
fitted_models = {}
fitted_scalers = {}

print("=== Fitting Final KMeans Models ===\n")
print(f"{'Dimension':<25} {'k':<5} {'N fitted':<12} "
      f"{'N zero':<10} {'Cluster sizes'}")
print("-" * 75)

for dim_name, k in k_final.items():
    features = feature_map[dim_name]
    label_col = f'cluster_{dim_name}'

    # Determine fitting population
    if dim_name in zero_config:
        zero_field, zero_label = zero_config[dim_name]
        active_mask = df_users[zero_field] > 0
        df_active = df_users[active_mask]
        n_zero = (~active_mask).sum()
    else:
        active_mask = pd.Series(
            [True] * len(df_users), index=df_users.index
        )
        df_active = df_users
        n_zero = 0

    # Scale
    X = df_active[features].fillna(0)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Fit
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE,
                n_init=N_INIT)
    labels = km.fit_predict(X_scaled)

    # Assign labels — zero population gets label 0
    # Active population gets labels 1 through k
    # (shift by 1 to reserve 0 for zero population)
    if dim_name in zero_config:
        df_users[label_col] = 0  # default = zero population
        df_users.loc[active_mask, label_col] = labels + 1
    else:
        df_users[label_col] = labels

    # Store model and scaler
    fitted_models[dim_name] = km
    fitted_scalers[dim_name] = scaler

    # Report
    sizes = pd.Series(
        df_users[label_col]
    ).value_counts().sort_index().values.tolist()
    print(f"{dim_name:<25} {k:<5} {len(df_active):<12,} "
          f"{n_zero:<10,} {sizes}")

# ── Verify ────────────────────────────────────────────────────────────
cluster_cols = [c for c in df_users.columns
                if c.startswith('cluster_')]
print(f"\n✓ Cluster columns added: {len(cluster_cols)}")
print(f"df_users shape: {df_users.shape}")

print("\n=== Cluster Label Counts ===\n")
for col in cluster_cols:
    print(f"{col}:")
    print(df_users[col].value_counts().sort_index()
          .to_string())
    print()
```

### This section shows findings from version 1 pipeline  
and is only intended as archived notes and lessons learned

- Session duration seconds  
    - needs bins for engagement tiers
    - keep 'is_timeout_session' flag - reduced to low priority  

- Base fare usd: use log1p to transform before clustering  
- page clicks bin tiers: bin: low (<13) / medium (13-43) / high (43-116) / power (116+)  
- clicks_per_second -> new feature: page_clicks / session_duration_sec (engagement intensity)  
- Booking lead time bins: last-minute (<7d), standard (7-30d), planners (30-100d), advanced (100+d)  

- origin_airport vs home_airport: binary flag: is_away_origin (non-home departure)
- destination distance groups bin: by geo distance from home airport (domestic/regional/international)
- hotel_nights vs trip_duration_days: delta feature: nights - trip_duration_days (flags hotel/flight mismatch)

**Users never booked**  
- never_booked users (432, 7.2%) = high browse, zero conversion  
- Likely perk: Exclusive Discounts (price barrier hypothesis)  
- These users will have NULL for ALL spend/flight/hotel features  
- Aggregation strategy: treat NULLs as zeros for spend, preserve browse metrics  

**Opportunity for retrospective **  
- sessions before 1st purchase
- where discounts involved with 1st purchase

**Correlation-based Feature Decisions**  
- *DROP* from clustering: `rooms` (redundant with seats and more deriviate that butts in `seats`, r=0.94)  
- *DROP* from clustering: `trip_duration_days` (redundant with `calc_nights`, r=0.98)  
- KEEP `page_clicks` AND `session_duration` despite r=0.85 (different signal types)  
- NOTE: `hotel_per_room_usd` is independent dimension - validate in hotel spending cluster  
- NOTE: discount amounts near-zero correlation with spend  
    - discount seekers are distinct behavioral type
